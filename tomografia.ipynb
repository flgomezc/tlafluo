```python
%pylab inline
from PIL import Image
import scipy.ndimage
```

```python
L = 3.0 # cm
l = 104 # px

def y_prim(n):
    return (n*2*L/l - L)
def x_prim(m):
    return (m*2*L/l - L)

def M(x): 
    return int((l/(2*L))*(x+L))
def N(y):
    return int((l/(2*L))*(y+L))

def X(xprim, yprim, theta):
    return xprim*cos(theta) - yprim*sin(theta)
def Y(xprim, yprim, theta):
    return xprim*sin(theta) + yprim*cos(theta)
```

```python
z = 20 #altura del plano de corte

Muestra = zeros([l,l])

# List of boxes to study
A = range(0,20)
#MCMC_reg = [open( '../results/%i.dat' %filenumber, 'w') for filenumber in A]
for filenumber in A:
    IM = Image.open('data/%i.tif' %filenumber)
    #figure()
    #imshow(IM)
    IMarray = array(IM)
    IMarray = IMarray /255
    I = (IMarray[z]-1)/255.0

    M_prim = zeros([l,l])
    theta = filenumber*pi/20
    for i in range(l):
        for j in range(l):
            AA = X(x_prim(i),y_prim(j),theta)
            BB = Y(x_prim(i),y_prim(j),theta)

            if (0<=M(AA))&(M(AA)<l):
                if (0<=N(BB))&(N(BB)<l):
                    M_prim[M(AA),N(BB)]+= I[j]
    #figure(figsize=(5,5))
    #imshow(M_prim)
    
    figure(figsize=(5,5))
    M2_prim = scipy.ndimage.filters.gaussian_filter(M_prim, 1.0)
    
    M2_prim[where(M2_prim>0.5)]=1.0
    M2_prim[where(M2_prim<0.5)]=0.0
    
    imshow(M2_prim)
    
    Muestra += M2_prim
    
imshow(Muestra)
```
